```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Sw=np.linspace(0,1,100) #devided into 100 parts
Krw=Sw**3;
Kro=(1-Sw)**3
plt.style.use('default')
plt.plot(Sw,Krw,label='Water Relative permeability curve') # Krw= relative permeability to water
plt.plot(Sw,Kro,label='Oil Relative permeability curve') # Kro= Relative pemeability to oil
plt.xlabel('Sw') ## Sw=water saturation
plt.ylabel('Krw')
plt.title('Relative Permeability plot') 
plt.grid()
plt.legend()
```
