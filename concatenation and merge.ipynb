### merging (to go or add with a unique column)

```python
test1=pd.DataFrame({'A':['X1','X2','X3','X4'],
'B':['Y1','Y2','Y3','Y4'], 'unique_ID':['Z1','Z2','Z3',
'Z5'],})
test1
```

```python
test2=pd.DataFrame({'C':['L1','L2','L3','L4'],
'D':['M1','M2','M3','M4'], 'unique_ID':['Z1','Z2','Z3','Z6']})
test2
```

```python
pd.merge(test1,test2, on='unique_ID', how='inner') # inner means intesection of both the data frame
```

```python
pd.merge(test1,test2, on='unique_ID', how='left') # going with 1st dataframe
```

```python
pd.merge(test1,test2, on='unique_ID', how='right') # going with the 2nd data frame
```

```python
pd.merge(test1,test2, on='unique_ID', how='outer') # outer means union of both the data frame
```

```python
test3=pd.DataFrame({'A':['X1','X2','X3','X4'],'B':['Y1','Y2','Y3','Y4'],'unique_ID_1':['Z1','Z2','Z5','Z6'],'unique_ID_2':['N1','N2','N6','N7']})
test3
```

```python
test4=pd.DataFrame({'C':['L1','L2','L3','L4'],'D':['M1','M2','M3','M4'],'unique_ID_1':['Z1','Z2','Z9','Z10'],'unique_ID_2':['N1','N2','N11','N12']})
test4
```

```python
pd.merge(test3,test4, on=['unique_ID_1','unique_ID_2'],how='inner')
```
