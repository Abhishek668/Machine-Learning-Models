```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_ng=pd.read_csv('monthly_csv.csv',index_col=['Month'])
df_ng.head()
```

```python
fig=plt.figure(figsize=(10,5), dpi=90)
plt.plot(df_ng.index.values,df_ng['Price'], color='red')
plt.xlabel('Date')
plt.ylabel('Gas Pricing, $/MMBTU')
plt.title('Gas Pricing Vs. Date')
```
